STEP 3. RFM METRICS

```python
from sklearn.preprocessing import StandardScaler 
from sklearn.cluster import KMeans
```

```python
last_date=df['Date'].max()
rfm=df.groupby('Date').agg({
    'Date':lambda x: (last_date-x.min()).days,
    'Unique.Visits': 'count',
    'Page.Loads': 'sum'
})
rfm.columns=['Recency','Frequency','Monetary']
rfm=rfm.reset_index()
scaler=StandardScaler()
rfm_scaled=scaler.fit_transform(rfm[['Recency','Frequency','Monetary']])
```

STEP 4. CUSTOMER SEGMENTATION

```python
kmeans=KMeans(n_clusters=4,random_state=42)
rfm['Cluster']=kmeans.fit_predict(rfm_scaled)
```

```python
plt.figure(figsize=(10,8))
scatter=plt.scatter(rfm['Recency'],rfm['Frequency'],c=rfm['Cluster'],cmap='rainbow')
plt.colorbar(scatter)
plt.xlabel('Recency')
plt.ylabel('Frequency')
plt.title('CUSTOMER SEGMENTATION BASED ON RFM ANALYSIS')
plt.savefig('rfm_Clusters.png')
plt.show()
```

```python
Cluster_avg=rfm.groupby('Cluster').mean()
print('Cluster Average:')
print(Cluster_avg)
```

```python
def assign_label(row):
  if row['Recency']<=7 and row['Frequency']>Cluster_avg['Frequency'].mean():
    return 'Active High Value'
  elif row['Recency']<=30:
    return 'Recent Customer'
  elif row['Frequency']>Cluster_avg['Frequency'].mean():
    return 'Loyal Customer'
  else:
    return 'At Risk'
rfm['Segment']=rfm.apply(assign_label,axis=1)
segment_counts=rfm['Segment'].value_counts()
plt.figure(figsize=(12,8))
segment_counts.plot(kind='bar')
plt.title('DISTRIBUTION OF CUSTOMER SEGMENTS')
plt.xlabel('Segment')
plt.ylabel('Number Of Customer')
plt.savefig('segment_distribution.png')
plt.show()   
```
